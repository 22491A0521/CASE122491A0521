```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

chandini=pd.read_csv('energy.csv')
print(chandini)

#line chart Andhra Karnataka
srh=chandini[chandini['State']=='Andhra']
rcb=chandini[chandini['State']=='Karnataka']

fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'])
diag[1].plot(rcb['Date'],rcb['Consumption'])
#bar chart Telengana Tripura
srh=chandini[chandini['State']=='Telengana']
rcb=chandini[chandini['State']=='Tripura']

fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'])
diag[1].bar(rcb['Date'],rcb['Consumption'])

#pie chart
cse4='Kerala'
cse5='Delhi'
tdp=[cse4,cse5]

srh=chandini[chandini['State']=='Kerala']['Consumption'].sum()
rcb=chandini[chandini['State']=='Kerala']['Consumption'].sum()
ysr=[srh,rcb]

fig,dig=plt.subplots(2)
diag[0].pie(ysr,labels=tdp)
```
