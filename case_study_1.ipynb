```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
harsha=pd.read_csv('energy.csv')
print(harsha)

srh=harsha[harsha['State']=='Andhra']
rcb=harsha[harsha['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='red')
diag[1].plot(rcb['Date'],rcb['Consumption'],color='blue')
srh=harsha[harsha['State']=='Telangana']
rcb=harsha[harsha['State']=='Tripura']
fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='red')
diag[1].bar(rcb['Date'],rcb['Consumption'],color='blue')
cse4='Kerala'
cse5='Delhi'
tdp=[cse4,cse5]
srh=harsha[harsha['State']=='Kerala']['Consumption'].sum()
rcb=harsha[harsha['State']=='Delhi']['Consumption'].sum()
ysr=[srh,rcb]
diag[0].pie(ysr,labels=tdp)
```
